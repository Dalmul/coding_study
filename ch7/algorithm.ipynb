### 다양한 정렬 알고리즘

- 정렬 알고리즘 : 주어진 목록을 정해진 순서대로 정렬
    - 버블 정렬 (Bubble Sort)
    - 삽입 정렬 (Insertion Sort)
    - 퀵 정렬 (Quick Sort)
    - 합병 정렬 (Merge Sort)

#### 정렬 알고리즘 : 버블 정렬

- 가장 간단하지만, 비효율적인 정렬 방식
1. 첫 번째 원소부터 시작하여, 인접한 두 원소를 비교
2. 앞의 원소가 뒤의 원소보다 크면 순서를 변경
3. 이 과정을 처음부터 끝까지 반복, 위의 과정으로 가장 큰 원소가 제일 오른쪽에 위치
4. 모든 원소가 정렬될 때까지 반복

#### 정렬 알고리즘 : 삽입 정렬

- 손에 들고 있는 카드를 정렬하는 것과 유사한 방식
1. 두 번째 원소부터 키(Key)로 지정
2. 키 앞 리스트에서 해당 키가 삽입될 위치를 탐색
3. 삽입될 위치를 찾으면, 키 보다 큰 원소들을 한 칸씩 밀어내고 해당 위치에 키를 삽입
4. 위의 과정을 모든 요소가 정렬될 때까지 반복

#### 정렬 알고리즘 : 퀵 정렬
- 분할 - 정복 알고리즘
1. 배열을 둘로 나눌 기준(피벗)을 결정
2. 피벗보다 작은 값들ㅇ르 왼쪽 배열에, 큰 값을 오른쪽 배열에 저장
3. 왼쪽 배열과 오른쪽 배열을 대상으로 위의 과정을 반복 수행
4. 왼쪽 배열 + 피벗 + 오른쪽 배열

## 대표적인 알고리즘

### - 검색 알고리즘
: 선형 검색 (Linear Search)

: 이진 검색 (Binary Search)

: 해시 테이블 (Hash Table)

### 무차별 대입(Brute Force)
- 문제의 모든 가능한 해를 생성하고, 계속하여 대입하여 해를 찾음
- 단순하고 직관적인 방법으로, 특별한 배경지식을 필요로 하지 않음
- 문제의 범위가 크거나 가능한 해의 종류가 많을 수록 효율이 떨어진다
- 모든 경우의 수를 고려

### 분할-정복 (Divide-and-Conquer)
- 큰 문제를 작은 하위 문제로 분할하고, 작은 문제를 해결 후 결합하여 큰 문제 해결
- 병렬로 큰 무제 해결 가능
- 분할 : 큰 문제를 더 이상 분할할 수 없을 때까지 재귀적으로 분할
- 정복 : 분할된 작은 문제를 해결
- 결합 : 해결된 하위 문제들을 원래 문제에 대한 결과로 조합

### 백트랙킹(Back Tracking)
- 주어진 문제를 임의의 방법으로 해결하다 실패하면 이전 상태로 돌악며 다른 방법으로 해결
- "답이 될 수 없는 상황"을 정의하고, 답이 될 수 없으면 끝까지 수행하지 않고 즉시 종료

### 동적 프로그래밍 (Dynamic Prigramming)
- 복잡한 문제를 해결하기 위한 문제해결 방버. 큰 문제를 여러개의 작은 문제로 나누어 해결
- 앞서 해결한 해를 재사용
- Memoization : 동일한 계산을 반복해야 할 때, 이전에 계산한 겂을 저장하여 재사용

### 재귀 (Recursion)
- 함수가 자기 자신을 호출하여 문제를 해결하는 알고리즘
- 재귀함수는 반드시 종료조건이 있어야 함
- 문제를 더 작은 단위로 나누고, 결과를 결합하여 문제를 해결
- ex) 피보나치 수열 : F(n) = F(n-1) + F(n-2)
- 자기 자신을 무한대로 호출할 경우, stackoverflow가 발생하기에 종료 조건 또는 입력값 제한 반드시 포함해야 함
